# Anomaly Detection for Electricity Theft using Isolation Forest

```python
import pandas as pd
from sklearn.ensemble import IsolationForest
```

```python
data = pd.read_csv(r"C:\Users\BVAYA\Desktop\energy-theft-prediction\data\processed\energy_featured.csv",parse_dates=['DateTime'],
                   index_col='DateTime')
```

```python
data.head()
```

```python
features = [
    'Global_active_power',
    'Global_reactive_power',
    'Voltage',
    'Global_intensity',
    'Rolling_mean_power',
    'Rolling_std_power',
    'Power_voltage_ratio',
    'Reactive_active_ratio',
    'Is_peak_hour',
    'is_weekend'
]

x = data[features]
```

```python
iso_model=IsolationForest(
    n_estimators=200,
    contamination=0.01,
    random_state=42,
    n_jobs=-1)
iso_model.fit(x)
```

Isolation Forest isolates anomalies instead of profiling normal behavior

```python
data['anomaly_label']=iso_model.predict(x)
```

```python
data['anomaly_label']=data['anomaly_label'].map({1:0,-1:1}) # conversion of binary
```

```python
data['anomaly_sccore']=-iso_model.score_samples(x)
```

Anomaly scores quantify how abnormal a consumption pattern is,
allowing prioritization instead of binary decisions.

```python
data['anomaly_label'].value_counts()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))

plt.plot(
    data.index,
    data['Global_active_power'],
    label='Power Consumption',
    alpha=0.6
)

plt.scatter(
    data[data['anomaly_label'] == 1].index,
    data[data['anomaly_label'] == 1]['Global_active_power'],
    color='red',
    label='Anomalies',
    s=10
)

plt.legend()
plt.title('Anomaly Detection in Power Consumption')
plt.show()
```

### Anomaly Detection Interpretation

Anomalies represent unusual electricity consumption patterns
that deviate significantly from historical behavior. These
patterns may indicate electricity theft, meter tampering,
or abnormal appliance usage and should be prioritized for inspection.
