```python
#Data Cleaning and Preprocessing
```

```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv(r"C:\Users\BVAYA\Desktop\energy-theft-prediction\data\raw\household_power_consumption.csv",sep=';')
```

```python
(data=='?').sum()
```

```python
# Missing values were represented using '?' symbols & converted to NaN for proper numerical processing
```

```python
data.replace('?',np.nan,inplace=True) 
```

```python
data.isna().sum()
```

```python
#All power, voltage, and sub-metering columns were converted to float
```

```python
numeric_cols=data.columns.drop(['Date','Time'])
data[numeric_cols]=data[numeric_cols].astype(float)   
```

```python
data.info()
```

```python
data.dropna(inplace=True)
```

```python
data.shape
```

```python
data['DateTime']=pd.to_datetime(data['Date']+' '+data['Time'],format='%d/%m/%Y %H:%M:%S')
```

```python
data.drop(['Date','Time'],axis=1,inplace=True)
```

```python
data.set_index('DateTime',inplace=True)
```

```python
data.sort_index(inplace=True)
```

```python
data.to_csv(r"C:\Users\BVAYA\Desktop\energy-theft-prediction\data\processed\energy_cleaned.csv")
```
