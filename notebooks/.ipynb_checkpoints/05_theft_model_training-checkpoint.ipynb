```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
import joblib
```

```python
data=pd.read_csv(r"C:\Users\BVAYA\Desktop\energy-theft-prediction\data\processed\energy_labeled.csv",parse_dates=['DateTime'],index_col='DateTime')
```

```python
data.head(2)
```

```python
Feature = [
    'Global_active_power',
    'Global_reactive_power',
    'Voltage',
    'Global_intensity',
    'Rolling_mean_power',
    'Rolling_std_power',
    'Power_voltage_ratio',
    'Reactive_active_ratio',
    'Is_peak_hour',
    'is_weekend'
]
x=data[Feature]
y=data['theft_label']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42,stratify=y)
# Stratify - used to preserve the class distribution during train test split
```

```python
model = RandomForestClassifier(n_estimators=200, # mony tress -better learning of patterns
                              random_state=42,  # same result every time
                              class_weight='balanced', # leanrs and protects rare theft cases 
                              n_jobs=-1  #  fastens traings - uses all cpu cores
                             )
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(accuracy_score(y_test,y_pred))
```

```python
cm = confusion_matrix(y_test,y_pred,labels=[0,1]) # test data has no theft cases s0 to balance the matrix we gave labels

plt.figure(figsize=(5,4))
sns.heatmap(cm,annot=True,fmt='d',cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')
plt.show()
```
