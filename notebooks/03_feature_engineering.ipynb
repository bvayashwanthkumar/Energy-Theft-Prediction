```python
# Feature Engineering for Energy Theft Detection
```

```python
import pandas as pd
```

```python
data=pd.read_csv(r"C:\Users\BVAYA\Desktop\energy-theft-prediction\data\processed\energy_cleaned.csv",parse_dates=['DateTime'],index_col='DateTime')
```

```python
data.head()
```

```python
#Create Time-Based Features
```

```python
data['Hour']=data.index.hour
data['Day']=data.index.day
data['Month']=data.index.month
data['Weekday']=data.index.weekday
data['is_weekend']=data['Weekday'].isin([5,6]).astype(int)
```

```python
#Power-Based Features
```

```python
data['Rolling_mean_power']=data['Global_active_power'].rolling(window=60).mean() #rolling means moving avg
data['Rolling_std_power'] = data['Global_active_power'].rolling(window=60).std()
```

```python
#Peak hour usage helps differentiate normal consumption from suspicious spikes.
```

```python
data['Is_peak_hour']=data['Hour'].between(18,22).astype(int)
```

```python
# Unusual power-to-voltage ratios may indicate meter bypass or tampering.
```

```python
data['Power_voltage_ratio'] = data['Global_active_power'] / data['Voltage']
```

```python
# High reactive power relative to active power can indicate inefficiency or abnormal load behavior.
```

```python
data['Reactive_active_ratio'] = data['Global_reactive_power'] / data['Global_active_power']
```

```python
data.dropna(inplace=True)
```

```python
data.isnull().sum()
```

```python
data.to_csv(r"C:\Users\BVAYA\Desktop\energy-theft-prediction\data\processed\energy_featured.csv")
```

```python
### Feature Engineering Summary

#- Temporal features capture time-based consumption patterns
#- Rolling statistics identify abnormal fluctuations
#- Power–voltage and reactive–active ratios help detect tampering
#- Engineered features enhance the predictive capability of ML models
```
